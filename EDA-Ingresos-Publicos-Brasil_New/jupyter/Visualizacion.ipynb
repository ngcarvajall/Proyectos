```python
df_llena_sinnull = pd.read_csv("../datos/df_llena_sinnull.csv", index_col=0, parse_dates=['DATA LANÇAMENTO'])
df_llena_sinnull.reset_index(drop=True, inplace=True)
```

```python
df_prueba_dif = df_llena_sinnull.groupby('CATEGORIA ECONÔMICA')[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO']].sum().reset_index()
df_prueba_dif['Diferencia'] = df_prueba_dif['VALOR REALIZADO'] - df_prueba_dif['VALOR PREVISTO ATUALIZADO']
df_prueba_dif ### se ha recibido menos o más
# 1. subejecucion, porque se ha recibido menos de lo previsto
# 2. sobreejecucion, porque se ha recibido más de lo previsto
# 3. subejecucion, porque se ha recibido menos de lo previsto
# 4. sobrejecucion, porque se ha recibido más de lo previsto
# 5. subejecucion, porque se ha recibido menos de lo previsto
```

```python
df_prueba_time1 = df_llena_sinnull.groupby(['ANO EXERCÍCIO', 'CATEGORIA ECONÔMICA'])[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO']].sum().reset_index()
df_prueba_time1
```

```python
### valores y su proporciones en base a los ingresos totales
ingresos_totales = df_llena_sinnull['VALOR REALIZADO'].sum()
ingresos_totales = ingresos_totales / 1000000000

df_llena_sinnull['VALOR PREVISTO ATUALIZADO'] = df_llena_sinnull['VALOR PREVISTO ATUALIZADO'] / 1000000000
df_llena_sinnull['VALOR LANÇADO'] = df_llena_sinnull['VALOR LANÇADO'] / 1000000000
df_llena_sinnull['VALOR REALIZADO'] = df_llena_sinnull['VALOR REALIZADO'] / 1000000000

df_llena_sinnull['Proporcion respecto a ingresos'] = (df_llena_sinnull['VALOR REALIZADO'] / ingresos_totales) * 100

dfprueba_proporcion= df_llena_sinnull.groupby("CATEGORIA ECONÔMICA")[["VALOR REALIZADO",'Proporcion respecto a ingresos']].sum()
dfprueba_proporcion
```
